Task-1:Generate a python program for a factorial of a number by using loops and without defineing any function.

```python
num = int(input("Enter a non-negative integer: ")) # Get input from user and convert to integer
factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,num + 1):
       factorial = factorial*i
   print("The factorial of",num,"is",factorial)
```

Task-3:Generate a python program for a factorial of a number by using user define function.

```python
def factorial(n):
  """Calculates the factorial of a non-negative integer."""
  if n < 0:
    return "Factorial does not exist for negative numbers"
  elif n == 0:
    return 1
  else:
    result = 1
    for i in range(1, n + 1):
      result *= i
    return result

# Get input from the user
num = int(input("Enter a non-negative integer: "))

# Calculate and print the factorial
print("The factorial of", num, "is", factorial(num))
```

Task-5:Generate a python function for both iterative and recursive
versions of the factorial function.

```python
def factorial_iterative(n):
  """Calculates the factorial of a non-negative integer iteratively."""
  if n < 0:
    return "Factorial does not exist for negative numbers"
  elif n == 0:
    return 1
  else:
    result = 1
    for i in range(1, n + 1):
      result *= i
    return result

def factorial_recursive(n):
  """Calculates the factorial of a non-negative integer recursively."""
  if n < 0:
    return "Factorial does not exist for negative numbers"
  elif n == 0:
    return 1
  else:
    return n * factorial_recursive(n - 1)

# Example usage:
num = 5
print(f"The iterative factorial of {num} is: {factorial_iterative(num)}")
print(f"The recursive factorial of {num} is: {factorial_recursive(num)}")
```
