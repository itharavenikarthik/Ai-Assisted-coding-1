Task-1: Generate a python function of a is_prime() function.

```python
def is_prime(n):
  """
  Checks if a number is prime.

  Args:
    n: An integer.

  Returns:
    True if n is prime, False otherwise.
  """
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

# Example usage
print(f"Is 7 prime {is_prime(7)}")
print(f"Is 10 prime {is_prime(10)}")
```

Task-2:Generate a python program for a function to reverse of string.

```python
def reverse_string(s):
  """
  Reverses a string.

  Args:
    s: The input string.

  Returns:
    The reversed string.
  """
  return s[::-1]

# Example usage
original_string = "hello"
reversed_string = reverse_string(original_string)
print(f"Original string: {original_string}")
print(f"Reversed string: {reversed_string}")
```

Task-3:Generate a python program for both recursive and iterative versions of a factorial function using comments.



```python
def factorial_recursive(n):
  """
  Calculates the factorial of a number recursively.

  Args:
    n: An integer.

  Returns:
    The factorial of n.
  """
  if n == 0:
    return 1
  else:
    return n * factorial_recursive(n-1)

def factorial_iterative(n):
  """
  Calculates the factorial of a number iteratively.

  Args:
    n: An integer.

  Returns:
    The factorial of n.
  """
  result = 1
  for i in range(1, n + 1):
    result *= i
  return result

# Example usage
num = 5
print(f"Factorial of {num} (recursive): {factorial_recursive(num)}")
print(f"Factorial of {num} (iterative): {factorial_iterative(num)}")
```

Task-4:Generate a python program to find the largest number in a list.



```python
def find_largest_number(numbers):
  """
  Finds the largest number in a list.

  Args:
    numbers: A list of numbers.

  Returns:
    The largest number in the list.
  """
  if not numbers:
    return None  # Return None for an empty list

  largest = numbers[0]
  for number in numbers:
    if number > largest:
      largest = number
  return largest

# Example usage
my_list = [10, 5, 23, 15, 8]
largest_number = find_largest_number(my_list)

if largest_number is not None:
  print(f"The largest number in the list is: {largest_number}")
else:
  print("The list is empty.")

my_empty_list = []
largest_number_empty = find_largest_number(my_empty_list)

if largest_number_empty is not None:
  print(f"The largest number in the empty list is: {largest_number_empty}")
else:
  print("Invalid syntax.")
```
